# Cleaning Orders Dataset

```python
try:
    df_orders = pd.read_csv('/olist_orders_dataset.csv')
    display(df_orders.head())
except FileNotFoundError:
    print("Error: 'olist_orders_dataset.csv' not found. Please upload the file or provide the correct path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df_orders.shape
```

```python
df_orders.dtypes
```

```python
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    df_orders[col] = pd.to_datetime(df_orders[col], errors='coerce')

display(df_orders.dtypes)
```

```python
df_orders.isnull().sum()
```

```python
df_orders['order_status'].value_counts()
```

```python
df_orders[df_orders['order_approved_at'].isnull()]['order_status'].value_counts()
```

```python
df_orders[(df_orders['order_approved_at'].isnull()) & (df_orders['order_status'] == 'delivered')]
```

```python
df_orders.loc[
    (df_orders['order_approved_at'].isnull()) &
    (df_orders['order_status'] == 'delivered'),
    'order_approved_at'
] = df_orders['order_purchase_timestamp']

df_orders.isnull().sum()
```

```python
print('Order status for missing order_delivered_carrier_date:')
display(df_orders[df_orders['order_delivered_carrier_date'].isnull()]['order_status'].value_counts())

print('\nOrder status for missing order_delivered_customer_date:')
display(df_orders[df_orders['order_delivered_customer_date'].isnull()]['order_status'].value_counts())
```

```python
# Impute missing order_delivered_carrier_date for 'delivered' orders
df_orders.loc[
    (df_orders['order_delivered_carrier_date'].isnull()) &
    (df_orders['order_status'] == 'delivered'),
    'order_delivered_carrier_date'
] = df_orders['order_approved_at']

# Impute missing order_delivered_customer_date for 'delivered' orders
df_orders.loc[
    (df_orders['order_delivered_customer_date'].isnull()) &
    (df_orders['order_status'] == 'delivered'),
    'order_delivered_customer_date'
] = df_orders['order_delivered_carrier_date']

display(df_orders.isnull().sum())
```

```python
inconsistent_delivery_data = df_orders[
    (df_orders['order_approved_at'].isnull()) &
    (
        (df_orders['order_delivered_carrier_date'].notnull()) |
        (df_orders['order_delivered_customer_date'].notnull())
    )
]

print(f"Number of inconsistent entries: {len(inconsistent_delivery_data)}")

if not inconsistent_delivery_data.empty:
    print("Inconsistent entries where order_approved_at is null but a delivery date is present:")
    display(inconsistent_delivery_data)
```

```python
inconsistent_carrier_delivery_data = df_orders[
    (df_orders['order_delivered_carrier_date'].isnull()) &
    (df_orders['order_delivered_customer_date'].notnull())
]

print(f"Number of inconsistent entries: {len(inconsistent_carrier_delivery_data)}")

if not inconsistent_carrier_delivery_data.empty:
    print("Inconsistent entries where order_delivered_carrier_date is null but order_delivered_customer_date is present:")
    display(inconsistent_carrier_delivery_data)
```

```python
df_orders.duplicated().sum()
```
